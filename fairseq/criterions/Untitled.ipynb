```python
import torch
import torch.nn as nn
```

```python
embed_dim = 512
kernel_size = 3
stride = 1
padding = 1
dilation = 1
conv = nn.Conv1d(embed_dim, embed_dim * 2, kernel_size, stride=stride, padding=padding, dilation=dilation)
```

```python
inputs = torch.randn(5, 512, 10)
```

```python
out = conv(inputs)
```

```python
out.shape
```

```python
out = out.transpose(1,2)
out.shape
```
